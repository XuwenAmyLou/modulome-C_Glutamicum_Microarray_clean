# 1_condition #
- parsed phase to phase, cultivation and growth rate

```python
condition_file = path.join('/Users/louxuwen/Desktop/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/4_unique_conditions/1_conditions_in.xlsx') # Enter metadata filename here
condition = pd.read_excel(condition_file,index_col=0).fillna("NaN")
```

```python
cultivations = condition["Cultivation"].tolist()
phase = condition["Phase"].tolist()
for i in range(len(phase)):
    temp = phase[i].split()
    if temp[0] == "after" or temp[0] == "in":
        cultivations[i] = phase[i]
        phase[i] = "NaN"
    else: 
        if len(temp) > 1:
            if temp[1][0].isdigit():
                s = ' '.join(temp[1:len(temp)])
                cultivations[i] = s
                phase[i] = temp[0]
            if temp[1] == "after" or temp[1] == "in":
                s = ' '.join(temp[1:len(temp)])
                cultivations[i] = s
                phase[i] = temp[0]
```

```python
condition["Cultivation"] = cultivations
condition["Phase"] = phase
```

```python
filepath = Path('/Users/louxuwen/Desktop/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/4_unique_conditions/1_conditions_out.xlsx')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
condition.to_excel(filepath) 
```

# 2_condition #
- parsed media

```python
condition_file = path.join('/Users/louxuwen/Desktop/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/4_unique_conditions/2_conditions_in.xlsx') # Enter metadata filename here
condition = pd.read_excel(condition_file,index_col=0).fillna("NaN")
```

```python
condition = condition.drop("glucose", axis=1)
```

```python
media = condition['media'].tolist()
media_type = []
component = []
for i in media:
    temp = i.split()
    if "CGXII" in i:
        media_type.append("CGXII")
        if len(temp) > 1:
            component.append(temp[temp.index("CGXII")+1])
        else:
            component.append("NaN")
    else:
        media_type.append("NaN")
        component.append("NaN")
```

```python
condition["media type"] = media_type
condition["component"] = component
```

```python
filepath = Path('/Users/louxuwen/Desktop/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/4_unique_conditions/2_conditions_out.xlsx')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
condition.to_excel(filepath) 
```
