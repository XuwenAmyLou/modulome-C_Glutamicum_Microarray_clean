# STEP 4: Dye Swap

only done once for the 525 experiments included in the net data file

```python
net_data_file = path.join('..','data','net_data','/home/amy/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/1_metadata_file/net_data.xlsx') # Enter metadata filename here
dye_file = path.join('..','data','dye_label_data','/home/amy/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/1_metadata_file/dye_label.xlsx') # Enter metadata filename here
net_data = pd.read_excel(net_data_file);
dye_label = pd.read_excel(dye_file);
```

```python
dye_label = dye_label.drop(columns = ['Unnamed: 2'])
dye_label_list = dye_label['Sample'].tolist()
swap = dye_label.loc[dye_label['Label'] == "Cy3/Cy5"]
swap_list_i = swap["Sample"].tolist()
print('Number of experiments that need to be swapped:', len(swap_list_i))
```

```python
sample_list = net_data["Sample name"].tolist()
swap_list = [i for i in swap_list_i if i in sample_list]
swap_row = []
for i in swap_list:
    rows = net_data.loc[net_data['Sample name'] == i]
    rows_num = rows['Unnamed: 0'].tolist()
    swap_row.extend(rows_num)
```

```python
net_data = net_data.drop(columns = ['Mean/Median','Fluorescence'])
```

```python
for i in swap_row:
    if net_data.at[i,'Label strain'] == 'Cy5':
        net_data.at[i,'Label strain'] = 'Cy3'
    else:
        net_data.at[i,'Label strain'] = 'Cy5'
```

```python
dye_data = net_data.copy()
filepath = Path('/home/amy/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/1_metadata_file/dye_data.xlsx')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
dye_data.to_excel(filepath) 
```

# STEP 5: Unique Conditions Sorting PART 1

```python
dye_data_file = path.join('..','data','net_data','/home/amy/Documents/GitHub/modulome-C_Glutamicum_Microarray_clean/1_metadata_file/dye_data_4.xlsx') # Enter metadata filename here
dye_data = pd.read_excel(dye_data_file);
dye_data = dye_data.drop(columns = ['Unnamed: 0','Unnamed: 0.1'])
```

```python
for i in range(0,dye_data.shape[0]):
    names = dye_data.at[i,'Sample name'].split(" vs. ")
    names[0] = names[0] + "-" + names[1][-4:]
    if dye_data.at[i,'Label strain'] == 'Cy5':
        dye_data.at[i,'Sample name'] = names[0]
    else:
        dye_data.at[i,'Sample name'] = names[1]
```
